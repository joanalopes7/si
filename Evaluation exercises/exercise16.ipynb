### Build, train and evaluate a neural network

**Instructions:**
- The training dataset has 32 features
- The task is binary classification (0,1)
- Use the SGD optimizer
- Use the BinaryCrossEntropy loss
- Use the accuracy metric
- The model should contain:


        Dense layer 1  |  ReLUactivation layer 1  |  Dense layer 2  |  ReLUactivation layer 2  |  Output Dense layer  |  Sigmoid activation layer


- **The dense layers should reduce the number of units to half (except the last one)**
- Train the NN for 100 epochs, with batchsize of 16 with a learning rate of 0.01.
- Test the model with k fold cross validation (hint: use the functions implemented in class 8)

```python
#Dataset and NN
import numpy as np
from si.data.dataset import Dataset
from si.neural_networks.neural_network import NeuralNetwork

#Optimizer
from si.neural_networks.optimizers import SGD

#Loss fucntion
from si.neural_networks.losses import BinaryCrossEntropy

#Model
from si.neural_networks.layers import DenseLayer
from si.metrics.accuracy import accuracy
from si.neural_networks.activation import ReLUActivation, SigmoidActivation

#Test
from si.model_selection.cross_validate import k_fold_cross_validation
```

```python
# Generate the dataset 
np.random.seed(42)
X = np.random.rand(1000, 32)  # 1000 samples, 32 features
y = np.random.randint(0, 2, size=(1000,))  # Binary labels (0,1)

dataset = Dataset(X, y)
```

```python
#Inspecting the dataset
data = dataset.to_dataframe()
print(f"The dataset have the following dimensions: {data.shape}")
print(f"Target unique values: {data['y'].unique()}")
data.head()
```

```python
# Define the model structure
net = NeuralNetwork(
    epochs=100, batch_size=16, optimizer=SGD, learning_rate=0.01, verbose=True, loss=BinaryCrossEntropy, metric=accuracy)
```

```python
# Adding nn layers 
n_features = dataset.X.shape[1]
net.add(DenseLayer(n_units=16, input_shape=(n_features,)))  # dense1
net.add(ReLUActivation())  #activation function 1
net.add(DenseLayer(n_units=8))  #dense2
net.add(ReLUActivation())  #activation function 1
net.add(DenseLayer(n_units=1))  #output dense layer
net.add(SigmoidActivation())  #output activation function 
```

```python
#Performing k fold cross validation
k = 5 
scores = k_fold_cross_validation(net, dataset, scoring=accuracy, cv=k, seed=42)
```

```python
# Results
print(f"Scores in each fold: {scores}")
print(f"Mean Accuracy: {np.mean(scores):.2f}")
```

The k-fold cross-validation returned an average accuracy of **95%**. This indicates that the model performs well on the given dataset, with a high capability to generalize to unseen data.
